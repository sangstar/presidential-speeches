```python
import pandas as pd
```

```python
df = pd.read_json('data/presidential-speeches.json')
```

```python
df['president'].value_counts()
```

```python
df = df.sample(frac=1)
```

```python
transcripts = df['transcript']
```

```python
import nltk
nltk.download('punkt')

sentences = [nltk.sent_tokenize(x.replace("\r\n\r\n"," ").replace("\n"," ").replace("\r","")) for x in transcripts]

sentences = [item for sublist in sentences for item in sublist]
```

```python
sentences
```

```python
import random 
random.shuffle(sentences)
```

```python
sentences
```

```python
train = pd.DataFrame(data = sentences)
```

```python
train.columns = ['text']
```

```python
train.to_json("data/train.jsonl",orient="records", lines = True)
```
